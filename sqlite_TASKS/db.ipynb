# SQLite Tasks

Use this notebook to quickly check the databases in this folder.

```python
import sqlite3
from pathlib import Path

base_dir = Path.cwd()
if not (base_dir / 'company_master.db').exists():
    base_dir = base_dir / 'sqlite_TASKS'

db_files = [
    base_dir / 'company_master.db',
    base_dir / 'project_records.db',
    base_dir / 'students.db',
]

for db_file in db_files:
    with sqlite3.connect(db_file) as conn:
        tables = conn.execute("SELECT name FROM sqlite_master WHERE type='table'").fetchall()
    print(db_file.name, tables)
```

```python
with sqlite3.connect(base_dir / 'students.db') as conn:
    conn.execute('''
        CREATE TABLE IF NOT EXISTS students(
            id INTEGER PRIMARY KEY AUTOINCREMENT,
            name TEXT,
            age INTEGER,
            marks INTEGER
        )
    ''')
    conn.commit()
```
