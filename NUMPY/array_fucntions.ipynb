ARRAY FUNCTIONS

```python
import numpy as np
a1=np.random.random((3,3))
a1=np.round(a1*100)
print(a1)
```

```python
#max/min/sum/prod

print(np.max(a1))
print(np.argmax(a1))
print(np.min(a1))
print(np.sum(a1))
print(np.prod(a1))
```

```python
#0---> col and 1---> rows

print(np.max(a1, axis=1))
print(np.max(a1,axis=0))
```

```python
#statistical operations

#medan/median/mode/std/var
print(np.mean(a1))
print(np.median(a1))
print(np.std(a1))
print(np.var(a1))
```

```python
#trigonometric functions
print(np.sin(a1))
print("-------------------------")
print(np.cos(a1))
print("-------------------------")
print(np.tan(a1))
```

```python
#dot peoduct

a2=np.arange(12).reshape(3,4)
a3=np.arange(12,24).reshape(4,3)

print(a2)
print("-------------------------")
print(a3)
print("-------------------------")
result = np.dot(a2, a3)
print(result)
```

```python
#LOG AND EXPONENETS

np.exp(a1)
```

```python
#floor/ceil/lround

print(np.floor(np.random.random((2,3))))
print("  ")
print(np.round(np.random.random((2,3))))
print("")
print(np.ceil(np.random.random((2,3))))
```
